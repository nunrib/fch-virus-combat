### challenge

- Make our fizzbuzz solution a function 
- Add a parameter for how what number we want to go to (instead of just 100)
- Add a parameter for where we want to start counting from

````
# Here is our previous solution for fizzbuzz. 
# How would we make it a function?
# Lets make it so we can pass a value to the range!

def fizzbuzz(start, count):
  for i in range(start,count+1):
    if i%3==0 and i%5==0:
        print("BuzzFizz")
    elif i%3==0:
        print("Fizz")
    elif i%5==0:
        print("Buzz")
    else:
        print(i)

````

````
# Call FizzBuzz
fizzbuzz(10, 20)
````

````
# Call FizzBuzz with different values again!
fizzbuzz(1,15)
````

````
# Call FizzBuzz with different values
fizzbuzz(1000, 1100)
````

### Variable length parameters

Sometimes you might want to be flexible with how many arguments can be passed in

Thats where we can use `*args` & `**Kwargs`

The important thing is the numer of `*` the rest can be a variable name you make


````
def names(*names):
  print(names)
  for name in names:
    print(name)
````

````
names("Ted",  "Jake", "Alexa", "bob", "Sue")
````

````
def person(**kwargs):
  print(kwargs)
  for key in kwargs:
    print(kwargs[key])
````

````
person(name = "Sage", city = "Seattle",  occupation = "Programmer")
````

## Scopes

Lets talk about how variables behave in functions

We're only going to focus on two types of scopes

`global` & `local`

just based on the names of these scopes do you have an intuition of how they might behave?



````
# variable declared out of a function (global)
name1 = "Sage"
print(name1)

# variable declared out of a function (global)
name1 = "Jake"
print(name1)
````

````
# global variable
name2 = "alexa"


def my_function():
  # print global variable insdie function
  print(name2)

my_function()

print(name2)
````

````
def my_function():
  # local variable made inside function
  global name3
  name3 = "Steve"
  print(name3)

my_function()

# print local variable froom function?
print(name3)

# how can we get it out?
# global
````

````
# Global Variable made outside function
name = "Sage"
print(name)

def my_function():
  # local variable made inside function
  name = "Jake"
  print(name)

my_function()

print(name)
````

````
# Global Variable made outside function
name = "Sage"
print(name)

def my_function():
  # local variable made inside function
  global name
  name = "Jake"
  return name

name = my_function()

print(name)
````

## Putting it together

Lets make some useful function we can reuse in our code.






````
# create function that has a parameter for a data object
# Verbose optional default is false
import pandas as pd
import numpy as np

def make_df(data_ob, verbose=False):

  # Create a dataFrame from the object passed in
  df = pd.DataFrame(data= np.c_[data_ob['data'], data_ob['target']],
                     columns= data_ob['feature_names'] + ['target'])
  
  # If verbose is True print information on load
  if verbose:
    print("Data Shape:")
    print(df.shape)
    print("\n---------------\n")
    print("Target Names:")
    print(data_ob['target_names'])
    print("\n---------------\n")
    print("Feature Names:")
    print(data_ob['feature_names'])
    print("\n---------------\n")
    print(df.describe())

  #return the DF out of function so we can use it later
  return df
````

Lets try our function out on the iris data set

````
from sklearn.datasets import load_iris
iris = load_iris()
````

````
iris
````

````
iris_df = make_df(iris)
````

````
iris_df
````

#Matplotlib

````
# We wont cover everything happening in depth in the cell 
# Read more about matplotlib at https://matplotlib.org/
import matplotlib.pyplot as plt
%matplotlib inline

# choose x & y index values. 
# Remember our list looks like this
# ['sepal length (cm)',
#  'sepal width (cm)',
#  'petal length (cm)',
#  'petal width (cm)']

# 0 = sepal length (cm)
# 3 = petal width (cm)
x_index = 1
y_index = 3

# this formatter will label the colorbar with the correct target names
# Look up lambda functions in python, they're neat!
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])


# figsize: changes size of chart (Width, Height)
# dpi: Dots per inch
plt.figure( figsize=(10, 10), dpi=90)

plt.scatter(iris.data[:, x_index], iris.data[:, y_index],
            c=iris.target, cmap=plt.cm.get_cmap('spring', 3))

plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
````
